```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('../data/airports.csv', dtype={'code': str,
                                                'brazilian': bool,
                                                'latitude': 'float32',
                                                'longitude': 'float32'})
df.head()
```

```python
px.set_mapbox_access_token(open('../scripts/.mapbox_token').read())
fig = px.scatter_mapbox(df, lat='latitude', lon='longitude', color='brazilian')
fig.update_layout(
    mapbox = { 'style': "mapbox://styles/aloysiogl/ckgs2cbsz10dj19rtt5zgox47", 'zoom': 0 },
    showlegend=False,
)
```
